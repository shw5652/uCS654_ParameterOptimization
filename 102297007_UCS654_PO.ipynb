```python
X=data.drop('Target', axis=1)
y=data['Target']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.svm import NuSVC
from sklearn.svm import *
```

```python
bAccuracy=0
bKernel=""
bNu=0
bEpsilon=0
```

```python
kernelList=['poly', 'sigmoid', 'rbf']
```

```python
def fitnessfunc(X_train, y_train, X_test, y_test, k, n, e):
    model = svm.NuSVC(kernel=k, nu=n)
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy
```

```python
for i in range(10):

    print('\n\nFor Sample ', i+1)

    bAccuracy=0
    bKernel=""
    bNu=0
    bEpsilon=0

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i*5)

    for j in range(100):
      selected_kernel = random.sample(kernelList, 1)[0]
      k = str(selected_kernel)
      print(k)
      n=round(float(random.uniform(0.0001,0.5)),5)
      print(n)
      e=random.random()
      print(e)
      accuracy=fitnessfunc(X_train, y_train, X_test, y_test, k, n, e)
      print('accuracy : ',accuracy)
      if accuracy > bAccuracy:
        bAccuracy=accuracy
        bKernel=k
        bNu=n
        bEpsilon=e

    print('Best Accuracy : ', bAccuracy)
    print('Best Kernel : ', bKernel)
    print('Best Nu : ', bNu)
    print('Best Epsilon : ', bEpsilon)
```
